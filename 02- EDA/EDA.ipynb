### 1 - Capitalizacion de Mercado
La capitalización del mercado es un indicador que mide y sigue el valor de mercado de una criptomoneda. En general, cuanto más alta sea la capitalización del mercado de una criptomoneda, más dominante se considerará esta en el mismo. Por esta razón, la capitalización del mercado suele considerarse el indicador más importante a la hora de clasificar las criptomonedas.

```python
mayor_cap=pd.read_csv("../00 - DATA\mayor_cap.csv")
```

```python
sns.set(rc={"figure.figsize":(12,4)})
ax=sns.barplot(x ='name', y ='market_cap',data=mayor_cap)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
ax. set (xlabel = ' Moneda / Coin',
       ylabel = ' Capitalizacion  ',
       title = ' Capitalizacion de Mercado ')
```

          En el grafico podemos observar las 10 monedas de mayor capitalizacion de mercado, y notamos que el Bitcoin esta capitalizada casi el doble que la segunda moneda mayor capitalizada.-

### Aca veremos como fueron capitalizadas las monedas a lo largo de 5 años

### 2 - Precio de la moneda

El valor de una criptomoneda cambia constantemente. El valor de una criptomoneda puede cambiar rápidamente, incluso cada hora. Y el monto de esa fluctuación puede ser considerable. Su valor depende de muchos factores, incluyendo la oferta y la demanda. Las criptomonedas tienden a ser más volátiles que las inversiones tradicionales, como los bonos y acciones.

```python
Precio_Total=pd.read_csv("../00 - DATA/Precio_Total.csv")
Precio_Total['Fecha'] = pd.to_datetime(Precio_Total['Fecha'])
```

```python
sns.set(rc={"figure.figsize":(12,4)}) 

ax = sns.lineplot(x="Fecha", y="prices",
             hue="Moneda",
             data=Precio_Total)
ax. set (xlabel = 'Fecha',
       ylabel = 'Precio',
       title = ' Historico de Precio ')
```

Aca observamos que el precio de la moneda estan completamente sesgadas por el bitcoin a lo largo de la historia.

### 4 - Volumen 

El volumen, también conocido como “volumen de negociación”, es el número de tokens, contratos u otras unidades operadas en un mercado durante un período de tiempo específico.

```python
Volumen=pd.read_csv("volumen.csv")
Volumen['Fecha'] = pd.to_datetime(Volumen['Fecha'])
```

```python
sns.set(rc={"figure.figsize":(12,4)})

ax = sns.lineplot(x="Fecha", y="total_volumes",
             hue="Moneda",
             data=Volumen)
ax. set (xlabel = 'Fecha',
       ylabel = 'Volumen',
       title = ' Historico de Volumen ')
```

Aca vemos el Volumen en el transcurso de 5 años y el BTC siempre fue la moneda de intercambio de todas las monedas.-

### 3 - Exchanges

Un exchange de criptomonedas es la plataforma en la que se realizan los intercambios de estas a dinero fíat o a otras criptomonedas. En estas casas de cambio online es donde se genera el precio de mercado que finalmente marca el valor de las criptomonedas en base a la oferta y demanda.

```python
Exchance=pd.read_csv("Exchange.csv")
```

```python
sns.set(rc={"figure.figsize":(12,4)})
ax=sns.barplot(x ='name', y ='trade_volume_24h_btc_normalized',data=Exchance)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
ax. set (xlabel = ' Exchance ',
       ylabel = ' Volumen ',
       title = ' Volumen de movimientos en las ultimas 24 horas ')
```

Y aca estan los 5 Exchange que mayor volumen de mercado manejan.

Las exchanges juegan un papel fundamental en el mercado de las criptomonedas debido a que proporcionan un lugar para que los usuarios compren, vendan e intercambien criptomonedas, lo que ayuda a aumentar la liquidez del mercado.-

## KPI


**Las KPI selecionadas para los datos que extraje son las siguientes:**


1 - ROI

El ROI es la sigla en inglés de “Return On Investment”, cuya traducción al español es “retorno de la inversión”. Se trata de una de las principales métricas que se utilizan a día de hoy para calcular la rentabilidad de algunas inversiones, como una campaña o acción concreta. Y es que, hasta la aparición de este tipo de indicadores, los resultados eran poco objetivos y certeros.

2 - Índice de Fuerza Relativa (RSI)

Una interpretación tradicional del RSI dice que por encima de 70 un activo está sobrecomprado, mientras que por debajo de 30 está sobrevendido. Así, los valores extremos podrían indicar un inminente cambio de tendencia o 'pullback'. Aún así, sería mejor no considerar estos valores como señales directas de compra o venta.

3 - Media Móvil (MA)

Una media móvil suaviza la acción del precio al filtrar el 'ruido' del mercado y resaltar la dirección de la tendencia. Dado que se basa en datos del precio pasados, se considera un indicador atrasado.
