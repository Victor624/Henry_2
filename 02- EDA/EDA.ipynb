## Por que es inportante saber estos 4 puntos a la hora de pensar en cripto-monedas?

### 1 - Capitalizacion de Mercado
La capitalización del mercado es un indicador que mide y sigue el valor de mercado de una criptomoneda. En general, cuanto más alta sea la capitalización del mercado de una criptomoneda, más dominante se considerará esta en el mismo. Por esta razón, la capitalización del mercado suele considerarse el indicador más importante a la hora de clasificar las criptomonedas.

### 2 - Precio de la moneda

El valor de una criptomoneda cambia constantemente. El valor de una criptomoneda puede cambiar rápidamente, incluso cada hora. Y el monto de esa fluctuación puede ser considerable. Su valor depende de muchos factores, incluyendo la oferta y la demanda. Las criptomonedas tienden a ser más volátiles que las inversiones tradicionales, como los bonos y acciones.

### 3 - Exchanges

Un exchange de criptomonedas es la plataforma en la que se realizan los intercambios de estas a dinero fíat o a otras criptomonedas. En estas casas de cambio online es donde se genera el precio de mercado que finalmente marca el valor de las criptomonedas en base a la oferta y demanda.


### 4 - Volumen 

El volumen, también conocido como “volumen de negociación”, es el número de tokens, contratos u otras unidades operadas en un mercado durante un período de tiempo específico.

```python
mayor_cap=pd.read_csv("mayor_cap.csv")
```

```python
sns.set(rc={"figure.figsize":(12,4)})
ax=sns.barplot(x ='name', y ='market_cap',data=mayor_cap)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
ax. set (xlabel = ' Moneda / Coin',
       ylabel = ' Capitalizacion  ',
       title = ' Capitalizacion de Mercado ')
```

          En el grafico podemos observar las 10 monedas de mayor capitalizacion de mercado, y notamos que el Bitcoin esta capitalizada casi el doble que la segunda moneda mayor capitalizada.-
